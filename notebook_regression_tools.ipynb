## Linear Regression for Diabetes dataset - Lesson 1

Import needed libraries

matplotlib, numpy, datasets, linear_model, model_selection을 import 진행

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

Load the diabetes dataset, divided into `X` data and `y` features

예시 데이터로 당뇨병 환자의 데이터셋을 불러옴. 
총 데이터 : 442개, 10개 특성(
-age
-sex
-bmi
-bp
-s1
-s2
-s3
-s4
-s5
-s6
)

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

Select just one feature to target for this exercise

한 개의 특성을 선택하여 실습을 진행함.

```python
# Selecting the 3rd feature
X = X[:, 2]
print(X.shape)
```

```python
#Reshaping to get a 2D array
#2D array 형식으로 출력 진행
X = X.reshape(-1, 1)
print(X.shape)
print(X)
```

Split the training and test data for both `X` and `y`

test size를 0.33으로 지정하여 random selection을 진행하여 데이터를 뽑아냄.

근데 이 과정에서 random selection을 할 경우 모델의 데이터 학습을 진행시에는 
모델을 만들고 생성 시에는 정확도에 문제가 생길 수 있다.

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

Select the model and fit it with the training data

모델을 선택하여 데이터를 넣음

모델 : linear_model.LinearRegression()

```python
model = linear_model.LinearRegression()
# 모델에 데이터를 입힘
model.fit(X_train, y_train)
```

Use test data to predict a line

테스트 데이터를 활용하여 예측 선을 그림

```python
y_pred = model.predict(X_test)
```

Display the results in a plot

결과를 plot으로 출력함

```python
# scatter(산점도)를 통해 데이터의 분포를 나타냄
plt.scatter(X_test, y_test,  color='black')
# 모델이 예측한 선형을 그려냄
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```

```python
# test -> train으로 변경하여 다시 그림

y_pred = model.predict(X_train)

plt.scatter(X_train, y_train,  color='black')
plt.plot(X_train, y_pred, color='blue', linewidth=3)
plt.show()
```
